```python
def clasificar_zona(valor):
    if 'URBANA' in valor and 'RURAL' in valor:
        return 'MIXTA'
    elif 'URBANA' in valor:
        return 'URBANA'
    elif 'RURAL' in valor:
        return 'RURAL'
    else:
        return 'NO DILIGENCIADO'

df_acceso_educacion['zona_normalizada'] = df_acceso_educacion['zona'].apply(clasificar_zona).astype('category')

```

```python
# Función para ordenar y normalizar niveles
orden_niveles = ['PREESCOLAR', 'BASICA PRIMARIA', 'BASICA SECUNDARIA', 'MEDIA', 'PRIMERA INFANCIA']

def normalizar_niveles(niveles):
    if pd.isna(niveles) or "NO DILIGENCIADO" in niveles:
        return 'NO DILIGENCIADO'
    partes = [n.strip() for n in niveles.split(',')]
    partes_ordenadas = [nivel for nivel in orden_niveles if nivel in partes]
    return ','.join(partes_ordenadas)

df_acceso_educacion['niveles_normalizados'] = df_acceso_educacion['niveles'].apply(normalizar_niveles).astype('category')
```

```python
def normalizar_jornada(jornada):
    if pd.isna(jornada) or "NO DILIGENCIADO" in jornada:
        return 'NO DILIGENCIADO'
    
    jornada = jornada.upper()
    categorias = ['MANANA', 'TARDE', 'NOCTURNA', 'UNICA', 'COMPLETA', 'FIN DE SEMANA']
    encontrada = [j for j in categorias if j in jornada]
    
    if len(encontrada) == 1:
        j = encontrada[0]
        return 'UNICA/COMPLETA' if j in ['UNICA', 'COMPLETA'] else j
    elif len(encontrada) > 1:
        return 'MIXTA'
    else:
        return 'OTRO'

df_acceso_educacion['jornadas_normalizadas'] = df_acceso_educacion['jornadas'].apply(normalizar_jornada).astype('category')
```

```python
prestadores_publicos = [
    'OFICIAL', 'CONCESION', 'EDUCACION MISIONAL CONTRATADA',
    'REGIMEN ESPECIAL', 'MUNICIPIO'
]

def clasificar_naturaleza(valor):
    if valor in prestadores_publicos:
        return 'PUBLICA'
    else:
        return 'PRIVADA'

df_acceso_educacion['naturaleza_institucion'] = df_acceso_educacion['prestador_de_Servicio'].apply(clasificar_naturaleza).astype('category')
```

# Modelo Machine Learning

#### Predecir si una institución es PRIVADA o PÚBLICA (naturaleza_institucion)
#### usando como variables predictoras:zona_normalizada y calendario



```python
# Paso 1: Importar librerías necesarias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Paso 2: Crear copia del DataFrame con las columnas necesarias
df_modelo = df_acceso_educacion[['zona_normalizada', 'calendario', 'jornadas_normalizadas']].copy()
```

```python
# Paso 3: Codificar variables categóricas (Label Encoding)
le_calendario = LabelEncoder()
df_modelo['calendario_cod'] = le_calendario.fit_transform(df_modelo['calendario'])

le_jornada = LabelEncoder()
df_modelo['jornada_cod'] = le_jornada.fit_transform(df_modelo['jornadas_normalizadas'])

le_zona = LabelEncoder()
df_modelo['zona_cod'] = le_zona.fit_transform(df_modelo['zona_normalizada'])
```

```python
le_naturaleza = LabelEncoder()
df_modelo['naturaleza_cod'] = le_naturaleza.fit_transform(df_acceso_educacion['naturaleza_institucion'])
```

```python
# Paso 4: Separar variables predictoras (X) y objetivo (y)
X = df_modelo[['zona_cod', 'calendario_cod', 'jornada_cod']]
y = df_modelo['naturaleza_cod']  # Ahora sí: lo que queremos predecir es si es PRIVADA o PUBLICA
```

```python
# Paso 5: Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Paso 6: Entrenar modelo
modelo = LogisticRegression(max_iter=1000)
modelo.fit(X_train, y_train)
```

```python
# Paso 7: Hacer predicciones
y_pred = modelo.predict(X_test)
```

```python
# Paso 8: Evaluar el modelo
print("📊 Matriz de Confusión:")
print(confusion_matrix(y_test, y_pred))

print("\n📋 Reporte de Clasificación:")
print(classification_report(y_test, y_pred, target_names=le_naturaleza.classes_))
```

```python
# Predecir si una institución es pública o privada, según jornada, zona y calendario

zona_test = le_zona.transform(['URBANA'])                    
calendario_test = le_calendario.transform(['B'])           
jornada_test = le_jornada.transform(['UNICA/COMPLETA'])      

# Ejecutar predicción con las tres variables en el orden entrenado
resultado = modelo.predict([[zona_test[0], calendario_test[0], jornada_test[0]]])

# Mostrar resultado traducido: 'PRIVADA' o 'PUBLICA'
le_naturaleza.inverse_transform(resultado)
```

```python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.linear_model import LogisticRegression

# Crear modelo base
modelo_base = LogisticRegression(max_iter=1000)

# Usar el nombre correcto para tu versión de sklearn
modelo_calibrado = CalibratedClassifierCV(estimator=modelo_base, method='sigmoid', cv=5)

# Entrenar modelo calibrado
modelo_calibrado.fit(X_train, y_train)
```

```python
# Ver las predicciones como clase:
y_pred = modelo_calibrado.predict(X_test)
```

```python
# Ver la probabilidad calibrada de que una institución sea PRIVADA:

y_prob = modelo_calibrado.predict_proba(X_test)
```

```python
# Hacer una predicción para una institución con jornada única, calendario B, y zona urbana:

zona_test = le_zona.transform(['URBANA'])
calendario_test = le_calendario.transform(['A'])
jornada_test = le_jornada.transform(['NOCTURNA'])

entrada = [[zona_test[0], calendario_test[0], jornada_test[0]]]

resultado = modelo_calibrado.predict(entrada)
probabilidad = modelo_calibrado.predict_proba(entrada)

print("🔮 Predicción:", le_naturaleza.inverse_transform(resultado)[0])
print("📊 Probabilidad de ser PRIVADA:", round(probabilidad[0][le_naturaleza.transform(['PRIVADA'])[0]] * 100, 2), "%")
```

#### El modelo de Machine Learning desarrollado utilizó tres variables predictoras clave: la zona geográfica de la institución (urbana, rural o mixta), el calendario académico (A, B u otro), y la jornada escolar (única/completa, mañana, tarde, nocturna, etc.), con el fin de predecir si una institución educativa en Colombia es pública o privada.  Tras un proceso riguroso de limpieza, estandarización y codificación de datos, se entrenó un modelo de regresión logística, evaluado a través de una matriz de confusión y un reporte de clasificación. 

## Resultados clave:
#### El modelo alcanzó una precisión aproximada del 83%, lo que indica una capacidad sólida para distinguir entre instituciones públicas y privadas a partir de las tres variables analizadas.

#### En particular, el modelo predijo una alta probabilidad de que instituciones con calendario B, ubicadas en zonas urbanas y con jornada única/completa sean privadas, lo cual respalda la hipótesis planteada.

#### El modelo también fue capaz de predecir correctamente muchos casos de instituciones públicas, especialmente en contextos rurales, con calendario A y jornadas diferentes a la única.

## Conclusión general:

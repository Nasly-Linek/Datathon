```python
# Reiniciar todo: importar pandas y volver a cargar el archivo CSV desde cero
import pandas as pd

df_acceso_educacion = pd.read_csv("ESTABLECIMIENTOS_EDUCATIVOS-COLOMBIA_20250603.csv", sep=",", encoding="utf-8")
```

```python
df_acceso_educacion.head()
```

```python
# Contar valores nulos por columna
df_acceso_educacion.isna().sum()


```

```python
df_acceso_educacion = df_acceso_educacion.drop(columns=[
    'año', 'secretaria', 'direccion', 'telefono', 'nombre_Rector',
    'etnias', 'sector', 'genero', 'caracter', 'especialidad',
    'grados', 'modelos_Educativos', 'capacidades_Excepcionales',
    'discapacidades', 'idiomas', 'numero_de_Sedes', 'estado',
    'internado', 'estrato_Socio_Economico', 'correo_Electronico'
])
```

```python
# Contar valores nulos por columna
df_acceso_educacion.isna().sum()
```

```python
# Limpiar texto: mayúsculas, sin espacios extra, etc.
df_acceso_educacion['zona'] = df_acceso_educacion['zona'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['tipo_Establecimiento'] = df_acceso_educacion['tipo_Establecimiento'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['niveles'] = df_acceso_educacion['niveles'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['jornadas'] = df_acceso_educacion['jornadas'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['prestador_de_Servicio'] = df_acceso_educacion['prestador_de_Servicio'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['propiedad_Planta_Fisica'] = df_acceso_educacion['propiedad_Planta_Fisica'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['calendario'] = df_acceso_educacion['calendario'].astype(str).str.upper().str.strip().str.split().str.join(' ')
df_acceso_educacion['resguardo'] = df_acceso_educacion['resguardo'].astype(str).str.upper().str.strip().str.split().str.join(' ')
```

```python
# Paso previo para asegurar uniformidad
df_acceso_educacion['matricula_Contratada'] = df_acceso_educacion['matricula_Contratada'].astype(str).str.upper().str.strip()
df_acceso_educacion['matricula_Contratada'] = df_acceso_educacion['matricula_Contratada'].replace({'SÍ': 'SI'})
```

```python
# Mostrar los valores únicos presentes en la columna 'matricula_Contratada'
# Esto permite identificar inconsistencias, errores de escritura o variaciones antes de procesarla
df_acceso_educacion['matricula_Contratada'].unique()
```

```python
df_acceso_educacion['matricula_Contratada'] = df_acceso_educacion['matricula_Contratada'].map({
    'SI': True,
    'NO': False
})
```

```python
df_acceso_educacion.dtypes
```

```python
df_acceso_educacion['codigodepartamento'] = df_acceso_educacion['codigodepartamento'].astype('object')
df_acceso_educacion['nombredepartamento'] = df_acceso_educacion['nombredepartamento'].astype('object')
df_acceso_educacion['codigomunicipio'] = df_acceso_educacion['codigomunicipio'].astype('object')
df_acceso_educacion['nombremunicipio'] = df_acceso_educacion['nombremunicipio'].astype('object')
df_acceso_educacion['codigoestablecimiento'] = df_acceso_educacion['codigoestablecimiento'].astype('object')
df_acceso_educacion['nombreestablecimiento'] = df_acceso_educacion['nombreestablecimiento'].astype('object')
df_acceso_educacion['resguardo'] = df_acceso_educacion['resguardo'].astype('object')
```

```python
df_acceso_educacion.dtypes
```

```python
# Reemplazo de valores nulos con texto estandarizado por columna
columnas_a_reemplazar = ['zona', 'niveles', 'jornadas', 'resguardo']

for col in columnas_a_reemplazar:
    reemplazo = f'no_diligenciado_{col.lower()}'
    df_acceso_educacion[col] = df_acceso_educacion[col].fillna(reemplazo)
```

```python
# Conversión a tipo categórico solo para columnas específicas
columnas_categoricas = [
    'zona', 'niveles', 'jornadas',
    'prestador_de_Servicio', 'propiedad_Planta_Fisica',
    'matricula_Contratada', 'calendario'
]

for col in columnas_categoricas:
    df_acceso_educacion[col] = df_acceso_educacion[col].astype('category')
```

```python
# Contar nulos y vacíos por columna
nulos_y_vacios = pd.DataFrame({
    'Nulos (NaN)': df_acceso_educacion.isna().sum(),
    'Vacíos ("")': (df_acceso_educacion == '').sum()
})
nulos_y_vacios
```
